# Finger Angle Test Results Analysis

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### What values the angles should take for each pose

```python
pose1_joints = {
    "finger1": [0] * 150,
    "finger2": [0] * 150,
    "finger3": [0] * 150,
}

pose2_joints = {
    "finger1": [np.degrees(1.35)] * 150,
    "finger2": [np.degrees(1.35)] * 150,
    "finger3": [np.degrees(1.35)] * 150,
}

pose3_joints = {
    "finger1": [0] * 150,
    "finger2": [0] * 150,
    "finger3": [np.degrees(1.35)] * 150,
}

pose4_joints = {
    "finger1": [np.degrees(1.35)] * 150,
    "finger2": [0] * 150,
    "finger3": [0] * 150,
}

pose5_joints = {
    "finger1": [0] * 150,
    "finger2": [0] * 150,
    "finger3": [np.degrees(1.35)] * 150,
}

pose6_joints = {
    "finger1": [np.degrees(1.35)] * 150,
    "finger2": [np.degrees(1.35)] * 150,
    "finger3": [np.degrees(1.35)] * 150,
}

pose7_joints = {
    "finger1": [0] * 150,
    "finger2": [np.degrees(1.35)] * 150,
    "finger3": [np.degrees(1.35)] * 150,
}
```

```python
angle_order = {
    0: -90,
    1: -60,
    2: -30,
    3: 0,
    4: 30,
    5: 60,
    6: 90,
}

pose_joints = {
    1: pose1_joints,
    2: pose2_joints,
    3: pose3_joints,
    4: pose4_joints,
    5: pose5_joints,
    6: pose6_joints,
    7: pose7_joints,
}

pose_joints_dfs = {key: pd.DataFrame(pose_joints[key]) for key in pose_joints.keys()}
```

### Read and process each file

```python
num_fingers = 3
pose = 7

pose_files = []
for i in range(-90, 91, 30):
    file_path = f"test_data/{num_fingers}-fingers-pose{pose}-angle{i}.csv"
    pose_files.append(pd.read_csv(file_path))

for i in range(len(pose_files)):
    for col in pose_files[i].columns:
        pose_files[i][col] = np.rad2deg(pose_files[i][col])
        
pose_files[0]
```

### Mean angle of error for every joint in a given pose/rotation combination

```python
col_means = [pose_files[i] - pose_joints_dfs[pose] for i in range(len(pose_files))]

col_means = [abs(col_means[i].mean(axis=0)) for i in range(len(col_means))]
```

```python
x = [i for i in range(len(col_means[i].keys()))]

my_cmap = plt.get_cmap('plasma')
# rescale = lambda y: (y - np.min(y)) / (np.max(y) - np.min(y))
rescale = lambda y: y / 90

for i, means in enumerate(col_means):

    plt.figure().set_figwidth(15)
    mean_bar = plt.bar(x=x, height=means.values, color=my_cmap(rescale(means.values)))
    plt.bar_label(mean_bar, fmt="%0.2f")
    plt.xlabel("Joint Name")
    plt.ylabel("Absolute Average Angle of Error (Degrees)")
    plt.ylim(0, 90)
    plt.title(f"Absolute Average Angle of Error for all Controllable Joints: Pose {pose}, Angle: {angle_order[i]}")
    plt.xticks(x, means.keys())
    plt.show()
```

### Mean angle of error for every rotation in a given pose/joint combination

```python
joint_errors_per_angle = {key: [col_means[i][key] for i in range(len(col_means))] for key in pose_joints[pose].keys()}
```

```python
x = np.array([i for i in range(len(pose_files))])

for joint, errors in joint_errors_per_angle.items():
    plt.plot(errors)
    plt.xlabel("Rotation of Hand (Degrees)")
    plt.ylabel("Absolute Average Angle of Error (Degrees)")
    plt.ylim(0, 90)
    plt.title(f"Absolute Average Angle of Error for Pose {pose}, Joint {joint}")
    plt.xticks(x, -90 + x * 30)
    plt.show()
```

### Distribution of every joint's absolute angle error in a given pose/rotation combination

```python
for i in range(len(pose_files)):
    plt.figure().set_figwidth(15)
    # sns.boxplot(abs(pose_files[i]))
    sns.boxplot(abs(pose_files[i] - pose_joints_dfs[pose]))
    plt.xlabel("Joint Name")
    plt.ylabel("Absolute Average Angle of Error (Degrees)")
    plt.title(f"Full Distributions of Angle of Error for all Controllable Joints: Pose {pose}, Angle: {angle_order[i]}")
    plt.show()
```

## Average Across all Poses

```python
all_pose_files = {i: [] for i in range(1, 8)}

for pose in range(1, 8):
    for angle in range(-90, 91, 30):
        file_path = f"test_data/{num_fingers}-fingers-pose{pose}-angle{angle}.csv"
        curr_df = pd.read_csv(file_path)
        for col in curr_df.columns:
            curr_df[col] = np.rad2deg(curr_df[col])
        all_pose_files[pose].append(curr_df)

all_col_means = {i: [] for i in range(1, 8)}

for current_pose in range(1, 8):
    for current_angle in range(len(all_pose_files[1])):

        current_col_means = all_pose_files[current_pose][current_angle] - pose_joints_dfs[current_pose]

        current_col_means = abs(current_col_means.mean(axis=0))
        all_col_means[current_pose].append(current_col_means)

mean_joint_errors_by_rotation = []
for current_angle in range(len(all_pose_files[1])):
    angles = all_col_means[1][current_angle]
    for current_pose in range(2, 8):
        angles += all_col_means[current_pose][current_angle]

    angles /= 7
    mean_joint_errors_by_rotation.append(angles)
```

```python
x = [i for i in range(len(mean_joint_errors_by_rotation[i].keys()))]

my_cmap = plt.get_cmap('plasma')
rescale = lambda y: y / 90

for i, means in enumerate(mean_joint_errors_by_rotation):

    plt.figure().set_figwidth(15)
    mean_bar = plt.bar(x=x, height=means.values, color=my_cmap(rescale(means.values)))
    plt.bar_label(mean_bar, fmt="%0.2f")
    plt.xlabel("Joint Name")
    plt.ylabel("Absolute Average Angle of Error (Degrees)")
    plt.ylim(0, 90)
    plt.title(f"Absolute Average Angle of Error Across all Poses for all Controllable Joints: Angle: {angle_order[i]}")
    plt.xticks(x, means.keys())
    plt.show()
```

```python
all_col_means_rotation = {i: [] for i in range(7)}

for pose in range(1, 8):
    running = 0
    for i in range(len(all_col_means[pose])):
        all_col_means_rotation[i].append(np.mean(all_col_means[pose][i]))

all_rotation_means = [np.mean(all_col_means_rotation[i]) for i in range(7)]
```

```python
x = np.array([i for i in range(len(all_rotation_means))])

plt.plot(all_rotation_means)
plt.xlabel("Rotation of Hand (Degrees)")
plt.ylabel("Absolute Average Angle of Error (Degrees)")
plt.title("Absolute Average Angle of Error of all Joints and all Poses")
plt.xticks(x, -90 + x * 30)
plt.show()
```
